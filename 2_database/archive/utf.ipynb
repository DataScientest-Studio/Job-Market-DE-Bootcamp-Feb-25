```python
import os
import pandas as pd
import psycopg2

def connect_db():
    """Establish connection to PostgreSQL"""
    try:
        conn = psycopg2.connect(
            dbname="adzuna", user="labor", password="labormarket", host="localhost", port=5432, options="-c client_encoding=UTF8"
        )
        # Debug PostgreSQL encoding
        cursor = conn.cursor()
        cursor.execute("SHOW server_encoding;")
        print("Server Encoding:", cursor.fetchone()[0])  # Expect "UTF8"
        
        cursor.execute("SHOW client_encoding;")
        print("Client Encoding:", cursor.fetchone()[0])  # Expect "UTF8"

        return conn
    except Exception as e:
        print(f"Error connecting to database: {e}")
        return None

def load_csv(file_path):
    """Load job categories from the CSV into a pandas DataFrame"""
    try:
        # Try reading the file with latin1 encoding
        df = pd.read_csv(file_path, header=0 ,encoding='utf-8')
        return df
    except Exception as e:
        print(f"Error reading CSV file: {e}")
        return None

csv_file_path = "../../data_collection/output_files/adzuna_category.csv"
df = load_csv(csv_file_path)
conn = connect_db()
cursor = conn.cursor()
# Define the insert query
insert_query = """
    INSERT INTO categories (tag, class, label)
    VALUES (%s, %s, %s);
"""
# Prepare job data
category_data = [
    (
        row["tag"],
        row["__CLASS__"],
        row["label"]
    )
    for _, row in df.iterrows()
]
cursor.executemany(insert_query, category_data)
conn.commit()
```
